```
import numpy as np
import pandas as pd
```

```
data = pd.read_csv('/data.csv')
```

```
#hasil encode fitur categorical
encoded_year = pd.get_dummies(data['year_created'], prefix='year')
encoded_month = pd.get_dummies(data['month_created'], prefix='month')
encoded_location = pd.get_dummies(data['location'],prefix='location')
encoded_verified = pd.get_dummies(data['verified'], prefix='verified')
encoded_days = pd.get_dummies(data['most_active_day'], prefix='most_active_day')
encoded_hour = pd.get_dummies(data['most_active_hour'], prefix='hour')
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numerical_data = data[['freq_of_most_hashtag','followers_count','friends_count','favourites_count','statuses_count']]
numerical_data = pd.DataFrame(scaler.fit_transform(numerical_data), columns=numerical_data.columns) #scale fitur numerik
categorical_data = pd.concat([encoded_hour, encoded_days, encoded_month, encoded_year, encoded_verified, encoded_location], axis=1)
```

```
preprocessed_data = pd.concat([numerical_data, categorical_data],axis=1)
label = data[['is_buzzer']]
```

```
preprocessed_data.to_csv('features.csv')
label.to_csv('label.csv')
```

```
from sklearn.feature_selection import mutual_info_classif

def get_categorical_idx(preprocessed_feat_names, categorical_feat_names):
  categorical_idx = []
  for idx,feat_name in enumerate(preprocessed_feat_names):
    if preprocessed_feat_names[idx] in categorical_feat_names:
      categorical_idx.append(idx)
  return categorical_idx


categorical_idx = get_categorical_idx(list(preprocessed_data.columns), list(categorical_data.columns))
mutual_information = mutual_info_classif(preprocessed_data, label, discrete_features=categorical_idx) #mendapatkan MI dari masing2 fitur hasil preprocessed

### mi is an array with the same shape as features
### each of mi element contain the value of mutual_information(feature, target)
# Driver code to test above 
# Python program for implementation of Quicksort Sort
  
# This function takes last element as pivot, places 
# the pivot element at its correct position in sorted 
# array, and places all smaller (smaller than pivot) 
# to left of pivot and all greater elements to right 
# of pivot 
def partition(mi,features,low,high): 
    i = ( low-1 )         # index of smaller element 
    pivot = mi[high]     # pivot 
  
    for j in range(low , high): 
  
        # If current element is smaller than or 
        # equal to pivot 
        if   mi[j] <= pivot: 
          
            # increment index of smaller element 
            i = i+1 
            mi[i],mi[j] = mi[j],mi[i]
            features[i], features[j] = features[j], features[i]
            
  
    mi[i+1],mi[high] = mi[high],mi[i+1]
    features[i+1],features[high] = features[high],features[i+1]
    return ( i+1 ) 
  
# The main function that implements QuickSort 
# arr[] --> Array to be sorted, 
# low  --> Starting index, 
# high  --> Ending index 
  
# Function to do Quick sort 
def quickSort(mi,features,low,high): 
    if low < high: 
  
        # pi is partitioning index, arr[p] is now 
        # at right place 
        pi = partition(mi,features,low,high) 
  
        # Separately sort elements before 
        # partition and after partition 
        quickSort(mi,features,low, pi-1) 
        quickSort(mi,features,pi+1, high)
  
# Driver code to test above 

def choose_k_highest(features, k,n): #memilih k fitur terpenting berdasarkan mutual information
    selected_feat = []
    for i in range(n-1,n-1-k,-1):
        selected_feat.append(features[i])
    return selected_feat

# YOUR CODE HERE
mutual_information = list(mutual_info_classif(preprocessed_data, label))
n = len(mutual_information) 
feature_names = list(preprocessed_data.columns)
quickSort(mutual_information,feature_names,0,n-1) 

def get_features(): #mengambil 5,10,15,20,25 fitur terpenting
  set_of_features = []
  for k in range(5,25,2):
    set_of_features.append(choose_k_highest(feature_names, k,n))
  return set_of_features

set_of_features = get_features()
banyak_fitur = [
    "5 Fitur Terpenting",
    "10 Fitur Terpenting",
    "15 Fitur Terpenting",
    "20 Fitur Terpenting",
    "25 Fitur Terpenting",
]
```
